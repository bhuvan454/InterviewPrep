```python
import os
import sys

sys.path.append("/workspaces/InterviewPrep/ML_algos")

import LogisticRegression as lr
from sklearn.model_selection import train_test_split
from sklearn import datasets

import matplotlib.pyplot as plt
```

```python
x, y = datasets.make_classification(n_samples=1000, n_features=4, n_classes=2, random_state=123)
print(x.shape, y.shape)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=123)
```

```python
model = lr.LogisticRegression()
model.fit(x_train, y_train)
predictions = model.predict(x_test)
accuracy = model.accuracy(y_test, predictions)

print(f"Accuracy: {accuracy}")
```

```python
y_preds = model.predict(x_test)
threshold = 0.5

y_classes = (y_preds > threshold).astype(int)

# Plotting
fig,ax = plt.subplots(1,1,figsize=(10,6))

ax.scatter(x_test[:,0], x_test[:,1], c=y_classes, cmap='coolwarm', alpha=0.7)
plt.show()
```
